# 📥 Carga de Recursos con DataLoader

La clase `DataLoader` permite administrar datasets, modelos y recursos descargados automáticamente.

## 🔧 Funciones principales
1. 📋 Listar datasets.
2. 🔍 Consultar información.
3. 📊 Cargar como NumPy.
4. 🐼 Cargar como DataFrame.
5. 🐼 Cargar como Dataset Tensorflow.

La clase **DataLoader** definida en **rna.data**, se encarga del manejo de todo lo asociado a la carga de datasets. Cada dataset se conoce con un nombre único. Algunas de las funciones que realiza son:
* **list_datasets**: listas los datasets disponibles en el repositorio git.
* **load_image_dataset**: carga dataset de imagenes en el repositorio git.
* **load_audio_dataset**: carga dataset de audio en el repositorio git.
* **load_dataframe**: carga un dataset como dataframe.
* **load_array**: carga un dataset como arreglo numpy.
* **dataset_info**: ofrece información del dataset (nombre, descripción, autores, url de descarga, descripión breve de atributos, cantidad de ejemplos, etc.) en formato **json**.
* **dataset_info_display**: muestra en pantalla la información del dataset en formato amigable.
* **dataset_directory**: directorio local donde se encuentra el dataset. Útil para realizar algun tipo de procesamiento para archivos individuales como audios e imágenes.

En el repositorio Git de datasets, cada conjunto de datos cuenta con su propia carpeta, nombrada según el dataset. En su interior se encuentran el archivo de datos (en formato .csv o .zip), un documento Markdown con la documentación detallada y un archivo de metadatos (info.json). Estos archivos se descargan localmente bajo demanda, organizándose en una subcarpeta con el mismo nombre del dataset.

## 🧮 Visualización de Audios en Grilla

Carga **lazy** (a demanda) de los audios, es decir, a medida que se los necesita, y los presenta utilizando las clases de visualización de `rna.visual`.

```python
from rna.visual.ClassAudioCard import AudioCard
from rna.visual.ClassGridCardViewer import GridCardViewer


# Parámetros
n_rows = 3
n_cols = 5
n_samples = n_rows * n_cols

# Cargar dataset
ds, class_names, metadata = DataLoader.load_audio_dataset("common_voice_test", shuffle=True)

print("Clases:", class_names)

sample_rate = metadata['sample_rate']  # Obtiene el sample rate del metadata detectado

# Crear las tarjetas
cards = []

for audio_tensor, label_tensor in ds.take(n_samples):

    audio = audio_tensor.numpy()
    label = int(label_tensor.numpy())

    cards.append(
        AudioCard(
            audio=audio,
            sample_rate=sample_rate,
            title=class_names[label],
        )
    )

# Mostrar
GridCardViewer(cards, cols=n_cols).show()
```

## ➕ Procesamiento de Audios

En esta sección se demuestra el uso de las funciones de procesamiento de audio disponibles en el módulo `rna.audio` para recortar silencios, ajustar la duración del audio, y calcular y transformar el espectrograma Mel en una imagen representativa.

```python
import matplotlib.pyplot as plt
import numpy as np
from rna.audio import recortar_silencios, ajustar_largo_audio, audio_a_espectrograma_mel, espectrograma_mel_a_imagen

# Tomamos un audio de ejemplo del dataset cargado en la sección anterior
for audio_tensor, label_tensor in ds.take(1):
    audio_original = audio_tensor.numpy()
    label = int(label_tensor.numpy())

# 1. Recortar silencios al inicio y al final
audio_sin_silencio = recortar_silencios(audio_original)

# 2. Ajustar la duración a 2 segundos
duracion_objetivo = 2.0  # segundos
audio_ajustado = ajustar_largo_audio(audio_sin_silencio, muestreo=sample_rate, tiempo_segs=duracion_objetivo)

# 3. Obtener espectrograma Mel (por ejemplo, con 64 coeficientes Mel)
n_mels = 64
espectrograma = audio_a_espectrograma_mel(audio_ajustado, muestreo=sample_rate, n_mels=n_mels)

# 4. Convertir espectrograma a imagen
imagen_espectrograma = espectrograma_mel_a_imagen(espectrograma)

# Calcular la duración del espectrograma en segundos
duration_espectrograma_s = len(audio_ajustado) / sample_rate

print(f"Audio original - Muestras........: {len(audio_original)}")
print(f"Audio sin silencios - Muestras...: {len(audio_sin_silencio)}")
print(f"Audio ajustado a {duracion_objetivo:4.2f}s - Muestras: {len(audio_ajustado)}")
print(f"Espectrograma Mel - Dimensiones..: {espectrograma.shape}")

# Graficar los resultados
fig, axes = plt.subplots(1, 2, figsize=(12, 4))

# Onda de audio ajustada
tiempo = np.arange(len(audio_ajustado)) / sample_rate
axes[0].plot(tiempo, audio_ajustado, color='b')
axes[0].set_title(f"Señal de Audio Ajustada ({class_names[label]})")
axes[0].set_xlabel("Tiempo (s)")
axes[0].set_ylabel("Amplitud")
axes[0].grid(True)

# Espectrograma Mel
# Usar 'extent' para establecer las coordenadas del eje x y 'origin' para la dirección del eje y
axes[1].imshow(imagen_espectrograma, cmap='inferno', aspect='auto',
                extent=[0, duration_espectrograma_s, 0, n_mels], origin='lower')
axes[1].set_title("Espectrograma Mel (Imagen)")
axes[1].set_xlabel("Tiempo (s)")
axes[1].set_ylabel("Bandas Mel")
axes[1].axis('on')

plt.tight_layout()
plt.show()
```
